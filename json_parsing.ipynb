```python
# 모듈 import
import json
import requests

url = 'https://apis.data.go.kr/B552584/UlfptcaAlarmInqireSvc/getUlfptcaAlarmInfo'
params = {'serviceKey':'SUXfVt013L4slaFREhvcXhfFzEULWvfmkVtMcMwBYUEgHWOF2x8X90hz/RAYNo8ODd0Y5RDDpvFIr1agV426WQ==', 'pageNo':'1', \
          'numOfRows':'100', 'year':'2020', 'returnType':'json'}
#서비스 키, 1페이지, 행의 개수, 년도, 데이터 타입 설정

response = requests.get(url, params = params)
print(response.text) #결과 확인. 해당 형태는 문자열 데이터
```

```python
response.json() #json() 함수는 우리가 받은 문자열 데이터를 json으로 출력함
```

```python
response.json()['response']['body']
#response키를 먼저 일고,
#response의 value 내에서 body를 key로 읽어옴
#이런 식으로 key-value 구조를 띄는 게 json
```

```python
#items 내 각 구별 데이터를 출력
items = response.json()['response']['body']['items']
items
#items 리스트를 출력하여 리스트 내에 json 포맷(key-value)을 확인
```

## for문을 활용하여 데이터를 담기

```python
import pandas as pd
#각 데이터를 담는 빈 리스트를 선언
districtName=[] #지역명
moveName=[] #권역명
issueVal=[] #발령농도
issueTime=[] #발령시간
issueDate=[] #발령일
clearDate=[] #해제일
issueGbn=[] #경보단계
itemCode=[] #미세먼지 항목 구분(PM10, PM25)

for item in items:
    districtName.append(item['districtName'])
    moveName.append(item['moveName'])
    issueVal.append(item['issueVal'])
    issueTime.append(item['issueTime'])
    issueDate.append(item['issueDate'])
    clearDate.append(item['clearDate'])
    issueGbn.append(item['issueGbn'])
    itemCode.append(item['itemCode'])
    
# 데이터 및 리스트를 데이터프레임으로 만들기
df = pd.concat([pd.DataFrame(districtName),
                    pd.DataFrame(moveName),
                    pd.DataFrame(issueVal),
                    pd.DataFrame(issueTime),
                    pd.DataFrame(issueDate),
                    pd.DataFrame(clearDate),
                    pd.DataFrame(issueGbn),
                    pd.DataFrame(itemCode)
                   ], axis = 1)
df.columns = ['지역명', '권역명', '발령농도', '발령시간', '발령일', '해제일', '경보단계', '미세먼지 항목 구분']
```

```python
df
```

```python
#excel 파일로 내보내기
#df.to_excel("미세먼지데이터.xlsx", index = False)
#csv 파일로 내보내기
df.to_csv("미세먼지데이터.csv", index = False)
```

## for문으로 year조정

```python
df_list = [2019, 2020, 2021, 2022]
df_list
```

```python
for i in df_list:
    url = 'https://apis.data.go.kr/B552584/UlfptcaAlarmInqireSvc/getUlfptcaAlarmInfo'
    params = {'serviceKey':'SUXfVt013L4slaFREhvcXhfFzEULWvfmkVtMcMwBYUEgHWOF2x8X90hz/RAYNo8ODd0Y5RDDpvFIr1agV426WQ==',\
              'year':str(i), 'numOfRows':'500', 'returnType':'json', 'pageNo':1}
    response = requests.get(url, params = params)
    df_temp = pd.json_normalize(response.json()['response']['body']['items'])
    df=df.append(df_temp)
```
